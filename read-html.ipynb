```python
from bs4 import BeautifulSoup

file = "/Users/joregan/Playing/librivox_mult/pg13028-images.html"
```

```python
with open(file, "r") as f:
    soup = BeautifulSoup(f, "html.parser")
```

```python
for element in soup.find_all("div", class_="pg_body_wrapper"):
    element.decompose()
```

```python
from bs4 import Tag

pieces = []

for child in soup.children:
    if isinstance(child, Tag) and child.name in ["h3", "p"]:
        pieces.append((child.name, child.text.replace("\n", " ").strip()))
```
