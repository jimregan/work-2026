```python
from bs4 import BeautifulSoup

file = "/Users/joregan/Playing/librivox_mult/pg13028-images.html"
```

```python
with open(file, "r") as f:
    soup = BeautifulSoup(f, "html.parser")
```

```python
for element in soup.find_all("div", class_="pg_body_wrapper"):
    element.decompose()
```

```python
from bs4 import Tag

pieces = []

for child in soup.children:
    if isinstance(child, Tag) and child.name in ["h3", "p"]:
        pieces.append((child.name, child.text.replace("\n", " ").strip()))
```

```python
import nltk
import os

nltk_data_path = os.path.expanduser("~/nltk_data")
nltk.download("punkt", download_dir=nltk_data_path)
nltk.download("punkt_tab", download_dir=nltk_data_path)  # needed in newer NLTK
nltk.data.path.append(nltk_data_path)
```

```python
for tag, text in pieces:
    sent_text = nltk.sent_tokenize(text, language='english')
    for sentence in sent_text:
        print(f"{tag}: {sentence}")
```
